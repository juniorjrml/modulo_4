```
import threading
#import urllib.request
import time
import random
from threading import Thread
```

```
# Uma função que execute um pouco mais devagar
def coisa_mais_lenta():
  #print("\|/")
  for i in range(1000):
    print(".", end="\t")
    #urllib.request.urlretrieve(imagePath, fileName)

# Função para executar como decorator e medir o tempo
def tempo_decorrido(acao):
  def wrapper(n):
    inicio = time.time()
    acao(n)
    fim = time.time()
    print("", end='\n')
    print("[{funcao}] Tempo total de execução: {tempo_total}".format(
            funcao=acao.__name__,
            tempo_total=str(fim - inicio), end='\n')
        )
  return wrapper


@tempo_decorrido # utilizando o decorator
def fazer_sequencial(n):
  for i in range(n):
    coisa_mais_lenta()

@tempo_decorrido # o mesmo ^
def usar_thread(n): # utilizanod threads para executar algo que e mais lento de forma concorrente
  threads = []
  for i in range(n):
    thread = threading.Thread(target=coisa_mais_lenta)
    threads.append(thread)
    thread.start()

  for i in threads:
    i.join()
```

```
n = 10 # passando o mesmo tamanho para manter uma "igualdade"
fazer_sequencial(n)
usar_thread(n)
```

```
contador = 10

def tarefa(nome_acao, valor_objetivo, passo): # tarefa que soma ou subtrai um valor ate que alcance um valor desejado 
  global contador
  while contador < valor_objetivo:
    contador += passo
    print("{} para {}".format(nome_acao, contador))
    tempo_espera = random.randint(0,3)
    time.sleep(tempo_espera)


def tarefa_a(n): # passando os parametros  para incrementar a variavel global
  tarefa("Incrementa", n, 1)


def tarefa_b(n): # passando para decrementar
  n2 = n * -1
  tarefa("Decrementa", n2, -1)


@tempo_decorrido
def executar_threads(n): #função usada para medir o tempo total gasto com o decorator
  thread1 = threading.Thread(target=tarefa_a, args=(n,))
  thread2 = threading.Thread(target=tarefa_b, args=(n,))

  thread1.start()
  thread2.start()

  thread1.join()
  thread2.join()


```

```
contador = 0
executar_threads(30)
```

<h1>Classe Thread sendo herdada</h1>

```
class MinhaClasseThread(Thread):
  def __init__(self, nome, thread_id, contador, processo):
    self.thread_id = thread_id
    self.nome = nome
    self.contador = contador
    super().__init__()

  # Chamanda quando é chamado o metodo start()
  def run(self):
    while(self.contador > 0):
      self.contador = processo(self.contador, self.nome, self.thread_id)

def processo(contador, nome, id):
  print("Id: {} | Nome: {} | Contador: {}".format(id, nome, contador-1))
  return contador-1

```

```
threads = []
threads.append(MinhaClasseThread("jao", 1, 150, processo))
threads.append(MinhaClasseThread("maria", 2, 100, processo))

for i in threads:
  i.start()

for i in threads:
  i.join()
```
